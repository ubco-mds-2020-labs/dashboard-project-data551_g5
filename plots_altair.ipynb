```python
selection = alt.selection_single()
country_chart = alt.Chart(combined11, title = 'Relationship between #Suicides and GDP_per_Capita').mark_point(filled = True, opacity = 0.6).encode(
    alt.X('gdp_per_capita:Q'),
    alt.Y('suicides_no'),
    alt.Size('population:Q', scale = alt.Scale(range=[0,2000])),
    alt.Order('population:Q', sort = 'ascending'),
    tooltip = [alt.Tooltip('country:N'),
               alt.Tooltip('suicides_no:Q'),
               alt.Tooltip('gdp_per_capita:Q')
              ],
    color = alt.condition(selection,'country', alt.value('darkgrey'), legend = None)).add_selection(selection)
country_chart
```
